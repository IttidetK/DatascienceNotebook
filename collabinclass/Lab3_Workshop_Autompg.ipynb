# Convert Data and Select atribute
MPG (miles per gallon), cylinders, engine displacement (cu. inches), horsepower, vehicle weight (lbs.), time to accelerate from O to 60 mph (sec.), model year (modulo 100), origin of car (1. American, 2. European,3-Asian)

Convert data

```python
#displacement
#Converting Cubic Inches to Cubic Centimeters  (1 cubic inch = 16.387 cubic centimeters)
data['displacement_in_cc']=data['displacement']*16.387
```

```python
#vehicle weight
#Convert lbs(Pound) to kg (1 pound (lb) is equal to 0.45359237 kilograms (kg).)
data['weight_in_kg']=data['weight']/2.205
```

```python
#mpg
#converting Mile per gallon to kilometers per liter ( 1MPG= 0.425143707 KPL)
data['mileage_kpl']=data['mpg']/2.352
```

# Modeling

Preparing Training and Testing data

```python
#Training data
X= newdata
X.head()
```

```python
#Target class
y=X['mileage_kpl']
y
```

```python
X=X.drop(['mileage_kpl'],axis=1)
X.head()
```

Preparing library for Modeling

```python
# !pip install sklearn
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
import math
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

# LinearRegression

```python
lr=LinearRegression() #สร้าง model

#Define number of Training and Testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
lr.fit(X_train,y_train)

print(X_train.shape, y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
print(X_test)
```

```python
y_pred=lr.predict(X_test)

print(X_test)
print(y_pred)
print(y_test)
```

```python
#mesurement
print('**** Linear Regression ****')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R2 Score:',metrics.r2_score(y_test, y_pred))
```

```python
print('score:', lr.score(X_test,y_test))
```

```python
plt.figure(figsize=(14, 8))
plt.title("Prediction Results using Linear Regression")
result = [x for x in range(80)]
plt.plot(result, y_test[:80], linewidth=3, label="Actual")
plt.plot(result, y_pred[:80], linewidth=3, label="Predicted")
plt.ylabel('MPG', size=15)
plt.xlabel('Instances', size=15)
plt.legend(fontsize=15)
plt.show()
```

```python
# predicting the mileage of a vehicle with the trained model to check whether the results are satisfactory.

#cylinders
#horsepower
#acceleration
#model year
#origin
#displacement_in_cc
#weight_in_kg


lr.predict([[8,100,9.7,90,1,1497,1160]])
```


# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
#print(X_train)
print(X_test.head())
```

```python
decisionTreeModel = DecisionTreeRegressor()
decisionTreeModel.fit(X_train, y_train)
```

```python
y_pred = decisionTreeModel.predict(X_test)
#print(y_pred)
```

Model Evaluation

```python
mse  = mean_squared_error(y_test,y_pred)
print(mse)
```

```python
rmse = np.sqrt(mse)
print(rmse)
```

```python
R2Score = r2_score(y_test,y_pred)
print(R2Score)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=30) #จำนวน Tree
rf.fit(X_train,y_train)
y_pred_rf = rf.predict(X_test)
```

```python
#print Mesurement
print('**** Random Forest ****')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_rf))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_rf))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_rf)))
print('R2 Score:',metrics.r2_score(y_test, y_pred_rf))
```

```python
#Plot Prediction Results using Random Forest from y_pred_rf
plt.figure(figsize=(14, 8))
plt.title("Prediction Results using Linear Regression")
result = [x for x in range(80)]
plt.plot(result, y_test[:80], linewidth=3, label="Actual")
plt.plot(result, y_pred_rf[:80], linewidth=3, label="Predicted")
plt.ylabel('MPG', size=15)
plt.xlabel('Instances', size=15)
plt.legend(fontsize=15)
plt.show()
```

# นายอิทธิเดช กระจ่าง 6430250440

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/wine500.csv')
X = df['description']
X
```

# CountVectorizer

```python
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)
```

```python
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

cosine_similarities = cosine_similarity(X_vec[0], X_vec).flatten()
most_similar_doc_index = np.argsort(cosine_similarities)[-2]

print(f"Most similar wine description :\n{df['description'][most_similar_doc_index]}")
print(f"\nCosine Similarity Score: {cosine_similarities[most_similar_doc_index]}")
```

```python
related_docs_indices = cosine_similarities.argsort()[::-1]
results = []
for i in related_docs_indices:
  results.append([i, cosine_similarities[i], df['description'][i]])

result_df = pd.DataFrame(results, columns=['Wine Index', 'Cosine Similarity', 'Description'])
result_df
```

# TfidfVectorizer

```python
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

tfidf_cosine_similarities = cosine_similarity(X_vec[0], X_vec).flatten()
most_tfidf_similar_doc_index = np.argsort(tfidf_cosine_similarities)[-2]

print(f"Most similar wine description :\n{df['description'][most_tfidf_similar_doc_index]}")
print(f"\nCosine Similarity Score: {tfidf_cosine_similarities[most_tfidf_similar_doc_index]}")
```

```python
related_docs_indices = tfidf_cosine_similarities.argsort()[::-1]
results = []
for i in related_docs_indices:
  results.append([i, tfidf_cosine_similarities[i], df['description'][i]])

result_df = pd.DataFrame(results, columns=['Wine Index', 'Cosine Similarity', 'Description'])
result_df
```
